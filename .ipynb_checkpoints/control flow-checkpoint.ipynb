```python
class PlayWithList():
    
    # constructor
    def __init__(self, data):
        self.data = data

    #method
    def max(self):
        max = 0
        for i in range(len(self.data)):
            if self.data[i] > self.data[max]:
                self.data[max] = self.data[i]

        return self.data[max]

    def min(self):
        min = 0
        for i in range(len(self.data)):
            if self.data[i] < self.data[min]:
                self.data[min] = self.data[i]

        return self.data[min]

    def range(self):
        max_value = self.max()
        min_value = self.min()
        return max_value, min_value, max_value - min_value
```

```python
def minmax(data):
    smallest = largest = data[0] # we are givining both smallest and largest to the initial value in the dataset

    for num in data:
        if num > largest:
            largest = num
        elif num < smallest:
            smallest = num

    return (smallest, largest)

g = [3,3,46,46,76,87,65,6,7,65,343,53,54]
minmax(g)
```

```python
def rev(data):
    '''The list passed in this function will be reversed'''
    return [num for num in range(len(data),0,-1)]

data = [1,2,3,4,5,6,7]
rev(data)
```

```python
def reverse_list(data):
    """Reverses a list of data in-place."""
    left, right = 0, len(data)-1
    while left < right:
        data[left], data[right] = data[right], data[left]
        left += 1
        right -= 1
    return data  # Return the reversed list

# Example usage
data = [1, 2, 3, 4, 5, 6, 7]
reverse_list(data)
```

```python
def oddproduct(data):
    '''Python function that takes a sequence of integer values and \n
    determines if there is a distinct pair of numbers in the sequence whose\n
    product is odd'''

    left, right = 0, len(data)-1
    while left < right:
        if (data[left]*data[right]) % 2 !=0:
            return True
        else:
            return False

        right -=1
        

data = [2,2,4,7,10,8, 9]
oddproduct(data)
```

```python
def has_distinct_odd_product(data):
    '''Determines if there is a distinct pair of numbers in the sequence whose
    product is odd.'''
    for i in range(len(data)):
        for j in range(i + 1, len(data)):
            product = data[i] * data[j]
            if product % 2 != 0:
                return True  # Found a pair with odd product
    return False  # No distinct pair with odd product found

# Example usage
numbers = [2, 2, 4, 10, 8, 9,4,6,8,2,2,7]
has_distinct_odd_product(numbers)
```

```python
def distinct_num(data):
    '''Python function that takes a sequence of numbers and determines
if all the numbers are different from each other'''
    for i in range(len(data)):
        for j in range(i + 1, len(data)):
            if data[i] == data[j]:
                return True
    return False

data = [1,5,3,4]
distinct_num(data)
```

```python
class MyClass:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        if isinstance(other, MyClass):
            return MyClass(self.value + other.value)
        elif isinstance(other, int):
            return MyClass(self.value + other)
        else:
            raise TypeError('Error type')

ad = MyClass('Abra')
sd = MyClass('ham')

adddd = ad + sd
adddd.value
```

```python
class Vector:
    """Represent a vector in a multidimensional space."""

    def __init__(self, d):
        """Create d-dimensional vector of zeros."""
        self.coords = [0] * d

    def __len__(self):
        """Return the dimension of the vector."""
        return len(self.coords)

    def __getitem__(self, j):
        """Return jth coordinate of the vector."""
        return self.coords[j]

    def __setitem__(self, j, val):
        """Set jth coordinate of vector to given value."""
        self.coords[j] = val

    def add(self, other):
        """Return the sum of two vectors."""
        if len(self) != len(other):  # relies on len method
            raise ValueError("Dimensions must agree")
        result = Vector(len(self))  # start with a vector of zeros
        for j in range(len(self)):
            result[j] = self[j] + other[j]
        return result

    def __eq__(self, other):
        """Return True if the vector has the same coordinates as the other."""
        return self.coords == other.coords

    def __ne__(self, other):
        """Return True if the vector differs from the other."""
        return not self == other  # rely on the existing eq definition

    def __str__(self):
        """Produce a string representation of the vector."""
        return '<' + str(self.coords)[1:-1] + '>'  # adapt list representation
```

```python
class SequenceIterator:
    def __init__(self, sequence):
        self.seq = sequence
        self.k = -1

    def __next__(self):
        self.k += 1
        if self.k < len(self.seq):
            return(self.seq[self.k])
        else:
            raise StopIteration()
            
    def __iter__(self):
        return self
```

```python
data = [1,654,545,6,563,4,5,6,7]
seq = SequenceIterator(data)

for element in seq:
    if element%2 == 0:
        print(element)
```

```python
class SequenceIterator:
    def __init__(self, sequence):
        self.seq = sequence
        self.k = 0

    def __next__(self):
        if self.k < len(self.seq):
            return(self.seq[self.k])
        else:
            raise StopIteration()
            
    def __iter__(self):
        return self
```

```python
data = [1,654,545,6,563,4,5,6,7]
seq = SequenceIterator(data)

for element in seq:
    if element%2 == 0:
        print(element)
```
