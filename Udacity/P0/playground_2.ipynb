```python
def longest_time_on_phone(calls):
    phone_times = {}

    for call in calls:
        phone1, phone2, date, duration = call
        duration_seconds = int(duration)

        # Update time for phone1
        phone_times[phone1] = phone_times.get(phone1, 0) + duration_seconds

        # Update time for phone2
        phone_times[phone2] = phone_times.get(phone2, 0) + duration_seconds

    # Find the telephone number with the longest time
    max_phone = max(phone_times, key=phone_times.get)
    max_time = phone_times[max_phone]

    return f"{max_phone} spent the longest time, {max_time} seconds, on the phone during September 2016."
    
longest_time_on_phone(calls)
```

```python
def possible_telemarketers(calls, texts):
    # Create sets of numbers for texts, incoming calls, outgoing calls
    text_numbers = set(text[0] for text in texts) | set(text[1] for text in texts)
    incoming_call_numbers = set(call[1] for call in calls)
    outgoing_call_numbers = set(call[0] for call in calls)

    # Identify possible telemarketers
    potential_telemarketers = outgoing_call_numbers - (text_numbers | incoming_call_numbers)

    # Convert the set to a sorted list
    sorted_telemarketers = sorted(potential_telemarketers)

    # Format the result message
    result_str = "These numbers could be telemarketers:\n" + "\n".join(sorted_telemarketers)

    return result_str

# Example usage
possible_telemarketers(calls, texts)

```

```python
def word_flipper(our_string):
    """
    Flip the individual words in a sentence

    Args:
       our_string(string): String with words to flip
    Returns:
       string: String with words flipped
    """
    
    # Split the sentence into words
    words = our_string.split()
    
    # Reverse each word
    flipped_words = [word[::-1] for word in words]
    
    # Join the reversed words back into a sentence
    flipped_sentence = ' '.join(flipped_words)
    
    return flipped_sentence

# Test the function
our_string = "Hello world"
result = word_flipper(our_string)
print(result)
```

**Example:**
* `arr = [0, 2, 3, 1, 4, 5, 3]`
* `output = 3` (because `3` is present twice)

The expected time complexity for this problem is `O(n)` and the expected space-complexity is `O(1)`.

```python
def duplicate_number(arr):
    """
    :param - array containing numbers in the range [0, len(arr) - 2]
    return - the number that is duplicate in the arr
    """
    n = len(arr)-1
    
    current = 0
    expected = 0
    
    for i in arr:
        current += i 
    for j in range(0,n):
        expected += j
    print(current, expected)

    x = current - expected
    return x
```

```python
def nth_row_pascal(n):
    if n == 0:
        return [1]

    current_row = [1]
    
    for i in range(1, n + 1):
        previous_row = current_row
        current_row = [1]
        
        for j in range(1, i):
            next_number = previous_row[j] + previous_row[j - 1]
            current_row.append(next_number)
            
            # Print details for each iteration of the inner loop
            print(f"Iteration {i}, Inner Loop: j={j}, previous_row={previous_row}, current_row={current_row}")

        current_row.append(1)
        
        # Print details for each iteration of the outer loop
        print(f"Iteration {i}, Outer Loop: current_row={current_row}")

    return current_row

# Example usage:
result_4 = nth_row_pascal(5)
print(f"Final Result: {result_4}")
```

```python
def pascal_tri(n):
    res = [[1]]

    for i in range(n - 1):
        temp = [0] + res[-1] + [0]
        row =[]

        for j in range(len(res[-1]) + 1):
            row.append(temp[j] + temp[j + 1])
        res.append(row)
    return res
```

```python
def pascal_tri_with_sum(n):
    res = [(1, 1)]  # Initialize with the first row and its sum

    for i in range(n - 1):
        temp = (0,) + res[-1][0] + (0,)
        row = []

        row_sum = 0  # Variable to store the sum of elements in the current row

        for j in range(len(res[-1][0]) + 1):
            element_sum = temp[j] + temp[j + 1]
            row.append(element_sum)
            row_sum += element_sum  # Accumulate the sum of elements in the current row

        res.append((tuple(row), row_sum))  # Store both the row and its sum in the result

    return res

# Example usage:
result_4_with_sum = pascal_tri_with_sum(4)
print(f"Result with Sum: {result_4_with_sum}")
```

```python
"""
Read file into texts and calls.
It's ok if you don't understand how to read files.
"""
import csv

with open('texts.csv', 'r') as f:
    reader = csv.reader(f)
    texts = list(reader)

with open('calls.csv', 'r') as f:
    reader = csv.reader(f)
    calls = list(reader)

"""
TASK 3:
(080) is the area code for fixed line telephones in Bangalore.
Fixed line numbers include parentheses, so Bangalore numbers
have the form (080)xxxxxxx.)

Part A: Find all of the area codes and mobile prefixes called by people
in Bangalore. In other words, the calls were initiated by "(080)" area code
to the following area codes and mobile prefixes:
 - Fixed lines start with an area code enclosed in brackets. The area
   codes vary in length but always begin with 0.
 - Mobile numbers have no parentheses, but have a space in the middle
   of the number to help readability. The prefix of a mobile number
   is its first four digits, and they always start with 7, 8 or 9.
 - Telemarketers' numbers have no parentheses or space, but they start
   with the area code 140.

Print the answer as part of a message:
"The numbers called by people in Bangalore have codes:"
 <list of codes>
The list of codes should be print out one per line in lexicographic order with no duplicates.

Part B: What percentage of calls from fixed lines in Bangalore are made
to fixed lines also in Bangalore? In other words, of all the calls made
from a number starting with "(080)", what percentage of these calls
were made to a number also starting with "(080)"?

Print the answer as a part of a message::
"<percentage> percent of calls from fixed lines in Bangalore are calls
to other fixed lines in Bangalore."
The percentage should have 2 decimal digits
"""

       # Part A
def get_codes_called_by_bangalore(calls):
    # Set to store unique codes
    codes_set = set()

    # Iterate through calls
    for call in calls:
        caller, receiver, _, _ = call
        
        # Check if the call is from Bangalore
        if caller.startswith('(080)'):
            # If the receiver is a fixed line, extract area code
            if receiver.startswith('('):
                area_code = receiver.split(')')[0][1:]
                codes_set.add(area_code)
            # If the receiver is a mobile number, extract prefix
            elif ' ' in receiver and (receiver[0] == '7' or receiver[0] == '8' or receiver[0] == '9'):
                prefix = receiver.split(' ')[0]
                codes_set.add(prefix)
            # If the receiver is a telemarketer number
            elif receiver.startswith('140'):
                codes_set.add('140')
    
    return sorted(list(codes_set))

lst = get_codes_called_by_bangalore(calls)

         # Part B
def percentage_of_calls_to_bangalore_fixed_lines(calls):
    # Counters for total calls from Bangalore and calls to Bangalore fixed lines
    total_calls_from_bangalore = 0
    calls_to_bangalore_fixed_lines = 0

    # Iterate through calls
    for call in calls:
        caller, receiver, _, _ = call

        # Check if the call is from Bangalore
        if caller.startswith('(080)'):
            total_calls_from_bangalore += 1

            # Check if the call is to a Bangalore fixed line
            if receiver.startswith('(080)'):
                calls_to_bangalore_fixed_lines += 1

    # Calculate the percentage
    if total_calls_from_bangalore > 0:
        percentage = (calls_to_bangalore_fixed_lines / total_calls_from_bangalore) * 100
        result_str = "{:.2f}%".format(percentage)
    else:
        result_str = "No calls from fixed lines in Bangalore."

    return result_str

result = percentage_of_calls_to_bangalore_fixed_lines(calls)

print(f"The numbers called by people in Bangalore have codes: {lst}")
print(f"{result} percent of calls from fixed lines in Bangalore are calls to other fixed lines in Bangalore.")
```
